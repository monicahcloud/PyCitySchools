# PyCity Schools Analysis

Overall, charter schools outperformed district schools in both reading and math. 
Charter schools have fewer students then District schools. Fewer students and a lower teacher to student ratio maybe the reason they out performed district schoosl.
Student budget spending per student is higher in bottom performing schools than top performing schools, however although they those schools received a higher budget per student, schools with a lower budget still out performing.


```python
#Import dependencies
import pandas as pd
```

```python
# Load in file
file1 = "../PyCitySchools/schoolsData.csv"
file2 = "../PyCitySchools/studentData.csv"
```

```python
#Read and display the CSV with Pandas
studentdata_df = pd.read_csv(file2)
studentdata_df.head()
```

```python
#Read and display the CSV with Pandas
schoolsdata_df = pd.read_csv(file1)
schoolsdata_df.head()

schoolsdata_df.rename(columns = {'name': 'school'}, inplace = True)
merged_df = studentdata_df.merge(schoolsdata_df, how = 'left', on = 'school_name')
```

```python
file3 = pd.merge(studentdata_df, schoolsdata_df, how = 'left', on = ['school_name', 'school_name'])
file3.head()
```

```python
#Read and display the CSV with Pandas
schoolsdata_df = pd.read_csv(file1)
schoolsdata_df.head()
```

```python
#Find Total Schools
#Find Total Students
total_schools = len(file3["school_name"].unique())
total_students = len(file3["Student ID"].unique())
```

```python
print(f'Total Schools: {total_schools}')
print(f'Total Students: {total_students}')
```

```python
#Find Total Budget
sum_budget = schoolsdata_df["budget"].sum()
sumbudget = "${:,}".format(sum_budget)
print(f'Total Budget: {sumbudget}')
```

```python
#Find Average Math Score
#Find Average Reading Score
average_math = file3["math_score"].mean()
average_math
average_reading = file3["reading_score"].mean()
average_reading

print(f'Average Math Scores: {average_math}')
print(f'Average Reading Scores: {average_reading}')
```

```python
# % Passing Math (The percentage of students that passed math)
# % Passing Reading (The percentage of students that passed reading)
passing_reading_count = file3[(file3["reading_score"]>=70)].count()["Student ID"]
passing_reading_count
```

```python
passing_math_score = file3[(file3["math_score"]>=70)].count()["Student ID"]
```

```python
passing_math_percentage = (passing_math_score / total_students) * 100
passing_math_percentage
```

```python
passing_reading_percentage = (passing_reading_count / total_students) * 100
passing_reading_percentage
```

```python
# % Overall Passing (The percentage of students that passed math **and** reading)
overall_count = file3[ (file3["math_score"]>=70)  & (file3["reading_score"]>=70)   ].count()["student_name"]
overall_count
```

```python
total_percentage = (overall_count / total_students) * 100
total_percentage
```

```python
#Create a Summary Table including School name, School Type, Total Students, Total School Budeget,
#Per Student Budget, Average Math Score, Average Reading Score,  
#% Passing Math (The percentage of students that passed math),
#% Passing Reading (The percentage of students that passed reading), 
#% Overall Passing (The percentage of students that passed math **and** reading) 

district_summary = pd.DataFrame({
    "Total Students ": [total_students],
    "Total Schools": [total_schools],
    "Total Budget" : sumbudget,
    "Average Math Scores" : average_math,
    "Average Reading Scores" : average_reading,
    "% Passing Math" : passing_math_percentage,
    "% Passing Reading" : passing_reading_percentage,
    "% Overall Passing" : total_percentage
    
    })

district_summary
```

```python
#Top Performing Schools (by overall passing)
by_school = file3.set_index('school_name').groupby(['school_name'])

by_school
```

```python
#school types
sch_types = schoolsdata_df.set_index('school_name')['type']
sch_types
```

```python
#find total students per school
stu_per_sch = by_school['Student ID'].count()

# school budget
sch_budget = schoolsdata_df.set_index('school_name')['budget']
```

```python
#find budget per student
stu_budget = schoolsdata_df.set_index('school_name')['budget']/schoolsdata_df.set_index('school_name')['size']
```

```python
#find average math score
avg_math = by_school['math_score'].mean()
avg_math.head()
```

```python
#find average reading score
avg_read = by_school['reading_score'].mean()
avg_read.head()
```

```python
passmath = file3[file3['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch*100
passmath.head()
```

```python
passreading = file3[file3['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch*100
passreading.head()
```

```python
# % Overall Passing (The percentage of students that passed math **and** reading)
overallpass = (passmath + passreading)/2
overallpass.head()
```

```python
school_summary = pd.DataFrame({
    "School Type": sch_types,
    "Total Students": stu_per_sch,
    "Per Student Budget": stu_budget,
    "Total School Budget": sch_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "Percentage Passing Math" : passmath,
    "Percentage Passing Reading" : passreading,
    "Percentage Overall Passing" : overallpass
   
})

school_summary.head()
```

```python
#Top Performing Schools (by overall passing)

top_performers = school_summary.sort_values(["Percentage Overall Passing"], ascending=[False])
top_performers.head()
```

```python
#Bottom Performing Schools (by overall passing)

bottom_performers = school_summary.sort_values(["Percentage Overall Passing"], ascending=[True])
bottom_performers.head
```

```python
grade_9th_df = file3.loc[file3["grade"] == "9th"].groupby(["school_name"])["math_score"].mean()
grade_10th_df = file3.loc[file3["grade"] == "10th"].groupby(["school_name"])["math_score"].mean()
grade_11th_df = file3.loc[file3["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
grade_12th_df = file3.loc[file3["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()
```

```python
# Combine the series into a dataframe
grade_summary_df = pd.DataFrame({"9th": grade_9th_df,
      "10th": grade_10th_df,
      "11th": grade_11th_df,
      "12th": grade_12th_df})
grade_summary_df[["9th", "10th", "11th", "12th"]]
```

```python
# create spending bins
bins = [0, 584.999, 614.999, 644.999, 999999]
group_name = ['< $585', "$585 - 614", "$615 - 644", "> $644"]
merged_df['spending_bins'] = pd.cut(merged_df['budget']/merged_df['size'], bins, labels = group_name)

#group by spending
by_spending = merged_df.groupby('spending_bins')

#calculations
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = merged_df[merged_df['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
# df build
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
})
#reorder columns
scores_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
scores_by_spend.index.name = "Per Student Budget"
scores_by_spend = scores_by_spend.reindex(group_name)
#formating
scores_by_spend.style.format({'Average Math Score': '{:.1f}',
                              'Average Reading Score': '{:.1f}',
                              '% Passing Math': '{:.1%}',
                              '% Passing Reading':'{:.1%}',
                              'Overall Passing Rate': '{:.1%}'})
```

```python
# create size bins
bins = [0, 999, 1999, 99999999999]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (>2000)"]
merged_df['size_bins'] = pd.cut(merged_df['size'], bins, labels = group_name)

#group by spending
by_size = merged_df.groupby('size_bins')

#calculations 
avg_math = by_size['math_score'].mean()
avg_read = by_size['math_score'].mean()
pass_math = merged_df[merged_df['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()

            
# df build            
scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
            
})
            
#reorder columns
scores_by_size = scores_by_size[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]

scores_by_size.index.name = "Total Students"
scores_by_size = scores_by_size.reindex(group_name)

#formating
scores_by_size.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```

```python
by_type = merged_df.groupby("type")

#calculations 
avg_math = by_type['math_score'].mean()
avg_read = by_type['math_score'].mean()
pass_math = merged_df[merged_df['math_score'] >= 70].groupby('type')['Student ID'].count()/by_type['Student ID'].count()
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('type')['Student ID'].count()/by_type['Student ID'].count()
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('type')['Student ID'].count()/by_type['Student ID'].count()

# df build            
scores_by_type = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall})
    
#reorder columns
scores_by_type = scores_by_type[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
scores_by_type.index.name = "Type of School"


#formating
scores_by_type.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```
